# TẠO BẢNG

```python
# Create the table in Snowflake if it doesn't exist
table_name = 'MP3_MASTER_SKU'
query_create_table = f"""
    CREATE TABLE IF NOT EXISTS MP3_DATA.{table_name} (
        "DATE_CREATE_SKU" VARCHAR(60),
        "BRISTLE" VARCHAR(60),
        "MACHINE_TYPE" VARCHAR(60),
        "VN_SKU" VARCHAR(60),
        "SUB_SKU" VARCHAR(60),
        "MDGM_SKU" VARCHAR(60),
        "DIVISION_OWNER" VARCHAR(60),
        "LOADING_METHOD" VARCHAR(60),
        "BUNDLE_NAME" VARCHAR(60),
        "MP_SKU_DESCRIPTION" VARCHAR(60),
        "CASE_COUNT_PCS" NUMBER(7,2),
        "CBM" NUMBER(15,9),
        "PALLET_MP" NUMBER(7,2),
        "CURRENT_STATUS" VARCHAR(60),
        "REPLACE_FOR_SKU" VARCHAR(60),
        "CODE_DATE" VARCHAR(60),
        "SOURCING_SUBS" VARCHAR(600),
        "SHORT_BUNDLE_NAME" VARCHAR(60),
        "PROJECT_NAME" VARCHAR(60)
    )
"""

with con.cursor() as cursor:
    cursor.execute(query_create_table)
```

# XỬ LÝ DATA

```python
df_sku = pd.read_csv(folder_path+sku_csv_name, skiprows=2, usecols=range(0,19))
df_sku.head()
```

```python
df_sku.columns = [
    "DATE_CREATE_SKU",
    "BRISTLE",
    "MACHINE_TYPE",
    "VN_SKU",
    "SUB_SKU",
    "MDGM_SKU",
    "DIVISION_OWNER",
    "LOADING_METHOD",
    "BUNDLE_NAME",
    "MP_SKU_DESCRIPTION",
    "CASE_COUNT_PCS",
    "CBM",
    "PALLET_MP",
    "CURRENT_STATUS",
    "REPLACE_FOR_SKU",
    "CODE_DATE",
    "SOURCING_SUBS",
    "SHORT_BUNDLE_NAME",
    "PROJECT_NAME"
]
df_sku.head()
```

```python
# Xóa dòng thừa, xóa dấu cách thừa nếu có
df_sku = df_sku.dropna(how='all')
df_sku = df_sku.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Xử lý các cột dạng số
df_sku['CASE_COUNT_PCS'] = pd.to_numeric(df_sku['CASE_COUNT_PCS'], errors='coerce')
df_sku['CBM'] = pd.to_numeric(df_sku['CBM'], errors='coerce')
df_sku['PALLET_MP'] = pd.to_numeric(df_sku['PALLET_MP'], errors='coerce')
```

```python
# Xử lý column Date Create SKU
df_sku['DATE_CREATE_SKU'] = df_sku['DATE_CREATE_SKU'].str[4:15]
df_sku['DATE_CREATE_SKU'] = pd.to_datetime(df_sku['DATE_CREATE_SKU'], errors='coerce')
df_sku['DATE_CREATE_SKU'] = df_sku['DATE_CREATE_SKU'].dt.strftime('%Y-%m-%d')
```
