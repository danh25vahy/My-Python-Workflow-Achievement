## Check APP Setup Matrix

```python
df_cap = pd.read_csv("O:\\Shared drives\\VEMP Supply Planning\\Production Planning\\PP Master data\\CAP Master Data\\Characteristics\\Z6CL_Material_class.xls",sep= "\t",skiprows=68,encoding = "utf-16",usecols=[1,4,7])

df_cap = df_cap[df_cap['Description'].notnull()]
df_cap = df_cap.drop('Description', axis=1)

df_cap = df_cap[df_cap['Material'] != 'Material']
df_cap = df_cap.drop('Material', axis=1)

df_cap = df_cap.rename(columns={'Characteristic':'Setup Matrix'})
df_cap = df_cap.loc[df_cap['Setup Matrix'].str.startswith(('INM', 'TE', 'PK')),:]
df_cap['Setup Matrix_cap']=df_cap['Setup Matrix']

df_cap = df_cap.dropna(how='all')
df_cap = df_cap.drop_duplicates()
df_cap = df_cap.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df_cap.tail(1)
```

```python
len(df_cap)
```

```python
df_app = pd.read_csv("O:\\Shared drives\\VEMP Supply Planning\\Production Planning\\PP Master data\\APP Master Data\\Setup Matrix\\APO Setup Matrix Generation PP.csv",usecols=[0])

df_app['Setup Matrix'] = df_app['Setup Matrix'].str[-8:]
df_app['Setup Matrix_app']=df_app['Setup Matrix']
df_app = df_app.loc[df_app['Setup Matrix'].str.startswith(('INM', 'TE', 'PK')),:]

df_app = df_app.dropna(how='all')
df_app = df_app.drop_duplicates()
df_app = df_app.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df_app.tail(1)
```

```python
len(df_app)
```

```python
#MERGE RIGHT
df_merge = df_app.merge(df_cap,how='right',on=['Setup Matrix'])
df_merge = df_merge.drop_duplicates()
df_merge = df_merge.loc[df_merge['Setup Matrix_app'].isnull(),:]
df_merge.head(1)
```

```python
# Append a blank row to prevent error divive by zero
df_merge = df_merge.append(pd.Series(), ignore_index=True)
```
